# 데이터 병합

```python
# 병합용 데이터 프레임 생성

df1 = pd.DataFrame(
    {
      '고객번호' : [1001, 1002, 1003, 1004, 1005, 1006, 1007],
      '이름' : ['강감찬', '홍길동', '이순신', '장보고', '유관순', '이율곡', '권율' ]
    },

    columns = ['고객번호', '이름']
)
```

```python
df1
```

```python
df2 = pd.DataFrame(
    {
      '고객번호' : [1001, 1001, 1005, 1006, 1008, 1001],
      '금액' : [10000, 20000, 15000, 5000, 100000, 30000 ]
    },

    columns = ['고객번호', '금액']
)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2)   # 교집합 병합
```

```python
# full join 방식은 키 값이 한쪽에만 았어도 데이터를 병합
pd.merge(df1, df2, how='outer') 
```

```python
# left(df1) key 가 있는 값은 모두 가져오는데 Value 가 없을 경우 NaN 채워진다
pd.merge(df1, df2, how='left') 
```

```python
# right(df2) key 가 있는 값은 모두 가져오는데 Value 가 없을 경우 NaN 채워진다
pd.merge(df1, df2, how='right') 
```

# 중복 데이터 처리

```python
# 1. 데이터 가져 오기
import pandas as pd

# 파일 경로를 찾고 변수 file_path에 저장
file_path='./data/bicycle_dup.csv'

# read_csv() 함수로 데이터프레임으로 변환
df = pd.read_csv(file_path, encoding='cp949')
df
```

## 중복 데이터 확인

```python
# duplicated() 메소드로 중복 데이터 찾기
# DataFrame.duplicated()는 True, False 의 boolean 형태의 Series로 반환

df.duplicated(['이용거리'])
```

```python
df.duplicated(['이용거리', '나이'])
```

```python
# keep='first' default 이며, 중복값이  있으면 첫번째 값을 duplicated 여부를 False로 반환하고 
# 나머지는 중복값에 대해서는 True 반환 

df.duplicated(['이용거리'], keep='first')  # 기본값
```

```python
# keep='last' 중복값이  있으면 첫번째 값을 duplicated 여부를 True로 반환하고 
# 나머지는 중복값에 대해서는 False 반환 

df.duplicated(['이용거리'], keep='last')  
```

```python
# keep=False 중복값을 모두 True로 반환 

df.duplicated(['이용거리'], keep=False)  
```

```python
df.drop_duplicates(['이용거리'], keep='first') 
```

```python
df.drop_duplicates(['이용거리'], keep='last') 
```

```python
df.drop_duplicates(['이용거리'], keep=False) 
```
