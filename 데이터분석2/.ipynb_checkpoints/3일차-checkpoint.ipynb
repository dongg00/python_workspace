## EDA(Exploratory Data Analysis, 탐색적 데이터 분석)
- 시각화를 통해 데이터 안의 유효한 패턴(인사이트)을 확인하느 단계
- 수집한 데이터를 관찰하고 이해하는 과정
- Business Understanding 단계에서 설정한 방향성과 목표에 부합하면 유효

## 데이터 품질
- 결측치, 일관성, 형태, 정보

## Data Preparation
- 분석 기법에 적합한 데이터를 편성하는 단계
- 모델링을 대비하는 단계
- 알고리즘에 넣을 데이터를 '적절하게' 준비
- feature-label split
- train-test split

## Machine Learning
- 지도(Supervised) 학습, 비지도 학습

## Modelling
- 다양한 modeling 기법과 알고리즘을 선택하고, 사용되는 파라미터들을 최적화하는 단계
- 모델의 성능을 평가하고, overfitting 되었는지 확인
- 세부적으로 Model Planning, Model Building으로 구분

## Planning(분석 모형 기획)
- 알고리즘 후보군 선정
- 속도, 예측력, 해석력을 고려해서 후보군 선정
- 내부 계산과정이 복ㅈ바할 수록 속도↓, 예측력↑, 해석력↓
- 일반적으로 기술통계 분야의 알고리즘은 속도와 해석력이 우수함 

## Building
- Planning에서 고른 알고리즘에 학습용 데이터로 학습하고, 시험용 데이터로 평가
- 평가 시 관점에 따라 평가지표가 달라짐
- case 1) 정답: 1, 예측: 2 -> 오차: 1
- case 2) 정답: 1001, 예측: 1002 -> 오차: 1
- 평가지표: 서로 다른 알고리즘 중 한 평가지표 기준으로 어떤 알고리즘의 정확도가 좋은가를 나타내는 척도

## Overfitting
- train 데이터에만 최적화되는 것
- test 데이터는 잘 구별하지 못하고 오히려 성능이 나빠지는 현상
- 얼마만큼 학습시키는 것이 좋은지 결정하는 것이 모델 설계에서 중요

## Evalution
- Modeling 결과가 프로젝트 목적에 부합하는지를 평가
- Modeling 과정, 분석 결과 등
- Communicate Results
- Executive Summary, Technical Report 작성

## Deployment
- 완성된 모덜을 실제 업무에 적용하기 위한 단계
- Operationalize
- 모니터링과 모델의 유지보수 계획 마련

```python
year = sample.split(' ')[0].split('-')[0]
month = sample.split(' ')[0].split('-')[1]
day = sample.split(' ')[0].split('-')[2]

hour = sample.split(' ')[1].split(':')[0]
minute = sample.split(' ')[1].split(':')[1]
second = sample.split(' ')[1].split(':')[2]

print(year, month, day, hour, minute, second)
```

```python
df['year'] = df['datetime'].apply(lambda x: int(x.split(' ')[0].split('-')[0]))
df['month'] = df['datetime'].apply(lambda x: int(x.split(' ')[0].split('-')[1]))
df['day'] = df['datetime'].apply(lambda x: int(x.split(' ')[0].split('-')[2]))

df['hour'] = df['datetime'].apply(lambda x: int(x.split(' ')[1].split(':')[0]))
df['minute'] = df['datetime'].apply(lambda x: int(x.split(' ')[1].split(':')[1]))
df['second'] = df['datetime'].apply(lambda x: int(x.split(' ')[1].split(':')[2]))
```

# feature selection
- 과적합을 방지하고, 일반화를 달성하기 위해 사용하는 방법
- filter, wrapper, embeded

# filter 
- 기초 통계량을 기반으로 feature 간의 관계에서 높은 상관계수(영향력)을 가지는 feature를 사용하는 방법
- 예) 표준편차가 0에 가까우면(해당 column에 저장된 epdlxjemfdl rjdml ehdflf) 해당 fature를 삭제
- filer 방식이 유효한 경우는 분야마다 차이가 있음

```python
df.describe().T.sort_values(by='std')['std']
```

# wrapper
- feature를 하나씩 넣고, 빼면서 평가지표 점수 확인 -> 시행착오를 거친다.
- forward Selection(전지 선택), Backward Elimination(후반 제거), stepwirse Selection(단계별 선택)

```python
featires = ['season', 'holiday', 'workingday', 'weather', 'temp', 'atemp',
       'humidity', 'windspeed', 'year', 'month', 'hour', 'dayofweek']
label = 'count'

# train-test split

train, test = df[0::2], df[1::2]
train, test = train.reset_index(), test.reset_index()

X_train, y_train = train[featires], train[label]
X_test, y_test = test[featires], test[label]

model = RandomForestRegressor()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
features = ['season', 'workingday', 'weather', 'temp', 'atemp',
       'humidity', 'windspeed', 'year', 'month', 'hour', 'dayofweek']
label = 'count'

# train-test split

train, test = df[0::2], df[1::2]
train, test = train.reset_index(), test.reset_index()

X_train, y_train = train[features], train[label]
X_test, y_test = test[features], test[label]

model = RandomForestRegressor()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.feature_selection import RFE   # Recursive Feature Elimination

selector = RFE(estimator=model)
selector.fit(X_train, y_train)
```

```python
selector.ranking_
```

```python
rfe_dt = pd.DataFrame()
rfe_dt['rank'] = selector.ranking_
rfe_dt['feature'] = features
```

```python
rfe_dt.sort_values(by='rank')
```
