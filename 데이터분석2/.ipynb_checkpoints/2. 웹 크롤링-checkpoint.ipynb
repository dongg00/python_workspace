### 웹 자동화

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
browser = webdriver.Chrome()   # Chrome driver 
```

```python
browser.get('http://naver.com')
```

```python
elem = browser.find_element(By.LINK_TEXT, '카페')    # 케페 찾아옴
print(elem)
```

```python
elem.get_attribute('href')
```

```python
elem.get_attribute('class')
```

```python
# 앞으로 가기
browser.forward()
```

```python
# 새로이 고팀
browser.refresh()
```

```python
elem = browser.find_element(By.ID, 'query')
```

```python
elem.send_keys('이선희')
```

```python
elem.send_keys(Keys.ENTER)
```

```python
browser.back()
```

```python
browser.get('http://naver.com')
elem = browser.find_element(By.ID, 'query')
elem.send_keys('이선희')
elem.send_keys(Keys.ENTER)
browser.back()
# 앞으로 가기
browser.forward()
```

```python
elems = browser.find_elements(By.TAG_NAME, 'a')
for e in elems:
    tag = e.get_attribute('href')
    print(tag)
```

```python
# daum 사이트로 이동
browser.get('http://daum.net')
```

```python
elem = browser.find_element(By.NAME, 'q')
```

```python
elem.send_keys('이선희')
```
